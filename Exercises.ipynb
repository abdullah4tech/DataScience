```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# Functional plot
x = np.linspace(1,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.title("Excersie")
plt.xlabel("X Label")
plt.ylabel("Y Label")
```

```python
# Object Oriented way
fig = plt.figure()
axes = fig.add_axes([1.0,1.0,1.0,1.0])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("OOP")
axes.plot(x,y)
```
