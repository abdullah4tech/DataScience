```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Students Attendance')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
```

```python
fig = plt.figure()

axies = fig.add_axes([0.1,0.1,0.8,0.8])
axies.set_xlabel('X Label')
axies.set_ylabel('Y Label')
axies.set_title('SOL Predictions')
axies.plot(x,y)
```

```python
fig = plt.figure()

axies1 = fig.add_axes([0.1,0.1,0.8,0.8])
axies2 = fig.add_axes([0.2,0.5,0.4,0.3])

axies1.plot(x, y)
axies1.set_title('Lager Plot')
axies2.plot(y,x)
axies2.set_title('Smaller Plot')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
fig.tight_layout()

axes[0].plot(x,y)
axes[0].set_title('First Table')

axes[1].plot(y,x)
axes[1].set_title('Second Table')
```

# Functional way of plotting

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X Lable')
plt.ylabel('Y Label')
plt.title('Functional')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r-')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

# Object Oriented way

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_title('First One')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')
axes[0].plot(x,y)
axes[1].set_title('First One')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
axes[1].plot(y,x, 'r-')

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
for current_axes in axes:
    current_axes.plot(x,y)
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,3))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('figure.png', dpi=200)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.set_title('My Title')
axes1.set_xlabel('X Label')
axes1.set_ylabel('y Label')
axes1.plot(x,x**2, label="boys")
axes1.plot(x,x**3, label="girls")

axes1.legend(loc=(0.1,0.1))
```
