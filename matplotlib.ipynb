```python
fig,axes = plt.subplots(nrows=1, ncols=2)

axes[0].set_title('First One')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')
axes[0].plot(x,y)
axes[1].set_title('First One')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
axes[1].plot(y,x, 'r-')

plt.tight_layout()
```

```python
def fibonacci(n):
    # Handle the first two numbers
    if n == 0:
        return 0
    if n == 1:
        return 1
    
    # Start with the first two numbers
    a = 0
    b = 1
    
    # Find Fibonacci numbers one by one
    for _ in range(2, n + 1):
        c = a + b  # Next number is the sum of the last two
        a = b      # Move 'b' to 'a'
        b = c      # Move 'c' to 'b'
    
    return b  # The final number is the answer

print(f"The Fibonacci number at position is {fibonacci(3)}")
```
