# Multi Index

```python
# How to use zip()
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1,2,3,1,2,3]
out = list(zip(outside, inside))
```

```python
out = pd.MultiIndex.from_tuples(out)
```

```python
out
```

```python
df = pd.DataFrame(randn(6,2), out, ['A', 'B'])
```

```python
df
```

```python
df.loc['G2'].iloc[1]
```

```python
df.index.names
```

```python
df.index.names = ['Group', 'Sub_Index']
```

```python
df
```

```python
df.loc['G1'].loc[2]['B']
```

## Groupby

#### Groupby allows you to group together rows based off a column  and perform an aggregate function on them.

```python
data = {
    'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
    'Person': ['Sam', 'Charles', 'Amy', 'Venessa', 'Carl', 'Sarah'],
    'Sales': [200, 130, 340, 125, 243, 350]
}
```

```python
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
g = df2.groupby('Company')
```

```python
g.min()
```

```python
g.sum()
```

```python
g.std
```

```python
g.count()
```

```python
g.min()
```
